```python
import pandas as pd

df = pd.read_excel("/content/BROWNWALL TEST FILE.xlsx")
df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    print(col)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)

print(df.columns)
```

```python
df = df.rename(columns={
    'company': 'client_name',
    'sales_person_name': 'salesperson',
    'est_order_value': 'expected_revenue',
    'timestamp': 'date'
})

print(df.columns)
```

```python
df['client_name'] = df['client_name'].astype(str).str.strip().str.upper()
df['salesperson'] = df['salesperson'].astype(str).str.strip().str.title()
```

```python
df['expected_revenue'] = pd.to_numeric(df['expected_revenue'], errors='coerce').fillna(0)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df = df.dropna(subset=['client_name', 'salesperson'])
df = df.drop_duplicates()
```

```python
df['is_sun_pharma'] = df['client_name'].str.contains('SUN PHARMA', na=False)
```

```python
df.to_excel("Cleaned_Brownwall_Data.xlsx", index=False)
```

```python
df.to_csv("Cleaned_Brownwall_Data.csv", index=False)
```
